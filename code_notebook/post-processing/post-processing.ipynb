```python
from sklearn.metrics import classification_report
import pandas as pd

def evaluate_classification(y_true, y_pred):

    # Calculate classification report
    report = classification_report(y_true, y_pred, output_dict=True)

    # Extract F1-score, accuracy, and precision from the report
    f1_score = report['weighted avg']['f1-score']
    accuracy = report['accuracy']
    precision = report['weighted avg']['precision']
    recall = report['weighted avg']['recall']

    # # Print the F1-score, accuracy, and precision
    # print("F1-score:", f1_score)
    # print("Accuracy:", accuracy)
    # print("Precision:", precision)
    # print("Recall:", recall)

    # Return the F1-score, accuracy, and precision as a dictionary
    return {'f1_score': f1_score, 'accuracy': accuracy, 'precision': precision, 'recall': recall}
```

```python
y_test_path = 'y_test.csv'
y_test = pd.read_csv(y_test_path)
```

### Llama2 

```python
test_path = 'y_test.csv'
predcition_path = './result-llama2/llama2-prediction_fewshot0.csv' #model name: LLAMA2 zero-shot
y_test = pd.read_csv(test_path)
prediction = pd.read_csv(predcition_path)
```

```python
evaluate_classification(y_test, prediction['prediction_result_encode'])
```

```python
predcition_path = './result-llama2/nhu_draft.csv' #model name: LLAMA2 fewshot5
prediction = pd.read_csv(predcition_path)
evaluate_classification(y_test, prediction['prediction_encode'])
```
